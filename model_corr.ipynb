```python
import pandas as pd
```

```python
building=pd.read_csv('building_data_corr.csv')
climate=pd.read_csv('climate_data_corr.csv')
target=pd.read_csv('target_labels_corr.csv')
```

```python
merged_df = pd.concat([building, climate, target], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
col_name = merged_df.select_dtypes(include='object').columns
for i in col_name:
    merged_df[i] = LabelEncoder().fit_transform(merged_df[i])
```

```python
merged_df.corr()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
X=merged_df.drop(['Climate_Impact'],axis=1)
y=merged_df["Climate_Impact"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.metrics import accuracy_score
lr.fit(X_train,y_train)
pred=lr.predict(X_test)
score=accuracy_score(y_test,pred)
```
